```python
# task_2
class Item:
    def __init__(self, name, count=0, max_count=100):
        self.name = name
        self.count = count
        self.max_count = max_count

    def __add__(self, num):
        """ Сложение с числом """
        return self.count + num

    def __mul__(self, num):
        """ Умножение на число """
        return self.count * num

    def __lt__(self, num):
        """ Сравнение меньше """
        return self.count < num

    def __len__(self):
        """ Длина объекта = текущее количество """
        return self.count
    
    def __eq__(self, value):
        return self.count == value

    def __lt__(self, value):
        return self.count < value

    def __le__(self, value):
        return self.count <= value

    def __ge__(self, value):
        return self.count >= value

        
    def __iadd__(self, num):
        self.count = min(max(self.count + num, 0), self.max_count)
        return self

    def __isub__(self, num):
        self.count = min(max(self.count - num, 0), self.max_count)
        return self

    def __imul__(self, num):
        self.count = min(max(self.count * num, 0), self.max_count)
        return self
    
    def __str__(self):
        return f"{self.name} count={self.count}"
```

```python
apple = Item("Apple", 5, 10)
print(apple)           
apple += 3
print(apple)       
apple += 5
print(apple)          
apple -= 20
print(apple)          
print(apple <= -1)
```

```python
# task_3
class No_Fruit(Item):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        
class Fruit(Item):
    def __init__(self, ripe=True, **kwargs):
        super().__init__(**kwargs)
        self._ripe = ripe


class Food(Item):
    def __init__(self, saturation, **kwargs):
        super().__init__(**kwargs)
        self._saturation = saturation
        
    @property
    def eatable(self):
        return self._saturation > 0
    
class Orange(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10, name = 'Orange'):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count, name = name)
        self._color = color
    
    @property
    def color(self):
        return self._color
    
class Mango(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10, name = 'Mango'):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count, name = name)
        self._color = color
    
    @property
    def color(self):
        return self._color
    
class Bread(Food, No_Fruit):
    def __init__(self, saturation=5, count=1, max_count=32, name="Bread"):
        super().__init__(name=name, saturation=saturation, count=count, max_count=max_count)


class Cheese(Food, No_Fruit):
    def __init__(self, saturation=5, count=1, max_count=32, name="Cheese"):
        super().__init__(name=name, saturation=saturation, count=count, max_count=max_count)
```

```python
# task_4
class Inventory:
    def __init__(self, size=10):
        self._slots = [None] * size
        self._size = size

    def __getitem__(self, index):
        return self._slots[index]

    def __setitem__(self, index, value):
        if not (0 <= index < self._size):
            raise IndexError("Индекс вне диапазона инвентаря")
        if value is not None and not isinstance(value, Item):
            raise TypeError("Можно добавлять только объекты Item")
        self._slots[index] = value

    def add(self, index, item):
        if not (0 <= index < self._size):
            raise IndexError("Индекс вне диапазона")
        if self._slots[index] is None:
            self._slots[index] = item
        elif isinstance(self._slots[index], type(item)):
            self._slots[index] += item._count
        else:
            raise ValueError("Ячейка занята другим предметом")

    def remove(self, index, count=1):
        if not (0 <= index < self._size):
            raise IndexError("Индекс вне диапазона")
        if self._slots[index] is None:
            raise ValueError("Ячейка пуста")
        self._slots[index] -= count
        if self._slots[index].count == 0:
            self._slots[index] = None

    def __str__(self):
        return str([str(obj) if obj else "Empty" for obj in self._slots])
    
```

```python
inv = Inventory(size=5)

orange = Orange(ripe=True, count=5)
bread = Bread(saturation=7, count=3)

inv.add(0, orange)
inv.add(1, bread)

print(inv)  

inv.remove(0, 3)
print(inv)  

inv.remove(0, 2)
print(inv)  
```

```python
# lab_1
class Data_Scientist:
    def __init__(self, name):
        self.name = name
        self._salary = 0
        
    def take_salary(self, amount):
        self._salary += amount
        
    @property
    def salary(self):
        return self._salary
    
    def read_matrix(self, filename):
        with open(filename, "r") as f:
            return [list(map(int, line.split())) for line in f]
        
class Pupa(Data_Scientist):
    def do_work(self, filename1, filename2):
        A = self.read_matrix(filename1)
        B = self.read_matrix(filename2)

        result = [[a + b for a, b in zip(row_a, row_b)] for row_a, row_b in zip(A, B)]
        
        for row in result:
            print(row)
        print(f'{self.name} выполнил задачу\n')
        
class Lupa(Data_Scientist):
    def do_work(self, filename1, filename2):
        A = self.read_matrix(filename1)
        B = self.read_matrix(filename2)

        result = [[a - b for a, b in zip(row_a, row_b)] for row_a, row_b in zip(A, B)]
        
        for row in result:
            print(row)
        print(f'{self.name} выполнил задачу\n')
        
class Accountant:
    def give_salary(self, worker: Data_Scientist, amount=1000):
        worker.take_salary(amount)
        print(f'{worker.name}: {worker.salary}')
```

```python
p = Pupa("Пупа")
l = Lupa("Лупа")
acc = Accountant()

p.do_work("m1.txt", "m2.txt")
l.do_work("m1.txt", "m2.txt")

acc.give_salary(p, 500)
acc.give_salary(l, 700)
```
