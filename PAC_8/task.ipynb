```python
# task_1
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

ohe = OneHotEncoder(sparse_output= False, handle_unknown= 'ignore')

df = pd.read_csv('data/wells_info_with_prod.csv')
df.drop(['PermitDate', 'SpudDate'], axis= 1, inplace= True)
Y = df['Prod1Year']
df.drop(['Prod1Year'], axis=1, inplace= True)

cat_col = ['CompletionDate', 'FirstProductionDate', 'operatorNameIHS', 'formation', 'BasinName', 'StateName', 'CountyName']
cat_data = df[cat_col]
df.drop(cat_col, inplace= True, axis= 1)
one_hot_data = ohe.fit_transform(cat_data)
one_hot_data = pd.DataFrame(one_hot_data, index=cat_data.index, columns=ohe.get_feature_names_out(cat_col))

X = pd.concat([df, one_hot_data], axis='columns')
```

```python
# task_2
X_train,  X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 42)
```

```python
# task_3&4
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

scaler_Y = StandardScaler()

Y_train_2d = Y_train.values.reshape(-1, 1)  
Y_test_2d = Y_test.values.reshape(-1, 1)    

Y_train_scaled = scaler_Y.fit_transform(Y_train_2d)
Y_test_scaled = scaler_Y.transform(Y_test_2d)

Y_train_scaled = Y_train_scaled.ravel()  
Y_test_scaled = Y_test_scaled.ravel()    
```
