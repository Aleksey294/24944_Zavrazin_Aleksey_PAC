```python
import cv2
import os
import numpy as np

images_path = "nails_segmentation/images"
masks_path = "nails_segmentation/labels"
```

```python
# task_1
image_files = sorted(os.listdir(images_path))
mask_files = sorted(os.listdir(masks_path))

pairs = []
for img_file in image_files:
    name_file = os.path.splitext(img_file)[0] + os.path.splitext(img_file)[1]

    if name_file in mask_files:
        pairs.append((os.path.join(images_path, img_file), os.path.join(masks_path, name_file)))

pairs
```

```python
# task_2
i = 0

while True:
    img_path, mask_path = pairs[i]
    
    image = cv2.imread(img_path)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    
    mask_bgr = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    combined = np.hstack([image, mask_bgr])
    cv2.imshow("result", combined)
    
    key = cv2.waitKey(0) & 0xFF
    if key == 27:
        break
    elif key == ord('d'):
        i = (i + 1) % len(pairs)
    elif key == ord('a'):
        i = (i - 1) % len(pairs)

cv2.destroyAllWindows()
```

```python
# task_3
i = 0
while True:
    img_path, mask_path = pairs[i]
    image = cv2.imread(img_path)
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

    mask_bgr = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    combined = np.hstack([mask_bgr, image])

    cv2.imshow("result", combined)

    key = cv2.waitKey(0) & 0xFF
    if key == 27:
        break
    elif key == ord('d'):
        i = (i + 1) % len(pairs)
    elif key == ord('a'):
        i = (i - 1) % len(pairs)

cv2.destroyAllWindows()
```

```python
# task_4
video = cv2.VideoCapture("videoplayback.mp4")

while True:
    _, frame = video.read()
    
    cv2.imshow('video', cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) )
    
    if cv2.waitKey(5) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
